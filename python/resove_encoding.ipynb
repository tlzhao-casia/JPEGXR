```python
fname = '../image/decode/decode.h'
```

```python
def convert_utf8(data):
    ret = ''

    start = 0
    size = len(data)
    while True:
        if start >= size: break

        end = start
        while end < size and data[end] >= 128: end += 1
        if end >= size: break
        try:
            ret += data[start: end + 1].decode('utf-8')
        except:
            ret += data[end: end + 1].decode('utf-8')
        start = end + 1

    return ret

with open(fname, 'rb') as f:
    data = f.read()
    ret = convert_utf8(data)
    print(ret)
```

```python
def convert_file_utf8(fname):
    with open(fname, 'rt') as f:
        assert (f.encoding.lower() == 'utf-8')

    need_write = False
    with open(fname, 'rb') as f:
        data = f.read()
        try:
            data.decode('utf-8')
        except:
            ret = convert_utf8(data)
            need_write = True
            # with open(fname, 'wt') as f:
            #     f.write(ret)
    if need_write:
        with open(fname, 'wt') as f:
            f.write(ret)

convert_file_utf8(fname)
```

```python
import os
def convert_all_c_files(root):
    for d in os.listdir(root):
        d = os.path.join(root, d)
        if os.path.isdir(d):
            convert_all_c_files(d)
        elif d.endswith('.c') or d.endswith('.h'):
            print(f"Converting {d}")
            is_utf8 = False
            with open(d) as f:
                if f.encoding.lower() == 'utf-8': is_utf8 = True
            convert_file_utf8(d)

convert_all_c_files('..')
```
